## Library

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

## Data Wrangling

### Gathering Data 

```python
df = pd.read_csv('Air_Quality_Shunyi.csv')
df = df.drop(columns=['No', 'RAIN', 'wd', 'station'])
df
```

### Assessing Data

#### Info Dataset

```python
df.info()
```

#### Pengecekan Ukuran Dataset

```python
df.shape
```

#### Pengecekan Missing Value

```python
df.isnull().sum()
```

#### Pengecekan Outliers

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Menghitung batas atas dan batas bawah untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengidentifikasi outlier
outliers = ((df < lower_bound) | (df > upper_bound)).sum()

print(outliers)
```

#### Menampilkan Outliers dalam Bentuk Boxplot

```python
# Membuat box plot untuk setiap kolom
plt.figure(figsize=(12, 6))  # Ukuran plot bisa disesuaikan
df.boxplot()
plt.title('Box Plot untuk Kolom Tanpa "wd" dan "station"')
plt.ylabel('Nilai')
plt.xticks(rotation=45)  # Melabeli sumbu x dengan rotasi teks
plt.show()
```

### Cleaning Data

#### Menangani Missing Value

```python
# Mengisi missing value dengan mean untuk kolom-kolom numerik
kolom_numerik = df.select_dtypes(include=['int', 'float'])
df[kolom_numerik.columns] = df[kolom_numerik.columns].fillna(df[kolom_numerik.columns].mean())

# Mengisi missing value dengan modus untuk kolom objek jika mode() tidak kosong
kolom_objek = df.select_dtypes(include=['object'])
if not kolom_objek.empty:
    df[kolom_objek.columns] = df[kolom_objek.columns].fillna(df[kolom_objek.columns].mode().iloc[0])
```

#### Pengecekan Jumlah Missing Value Setelah Penanganan

```python
df.isnull().sum()
```

#### Menangani Outliers

```python
# Menghapus baris yang mengandung outliers berdasarkan batas atas dan bawah
df_cleaned = df[(df >= lower_bound) & (df <= upper_bound)].dropna()

# Menampilkan DataFrame setelah menghapus outliers
print(df_cleaned)
```

#### Pengecekan Jumlah Outliers Setelah Penanganan

```python
# Menghitung jumlah outliers setelah penghapusan
outliers_sesudah = ((df_cleaned < lower_bound) | (df_cleaned > upper_bound)).sum()

# Menampilkan jumlah outliers setelah penghapusan
print(outliers_sesudah)
```

### Pertanyaan 1

#### Bagaimana kualitas udara di Shunyi berdasarkan PM2.5 setiap tahunnya?

```python
rata2_pm25_per_tahun = df.groupby('year')['PM2.5'].mean()

plt.figure(figsize=(10, 6))
plt.plot(rata2_pm25_per_tahun.index, rata2_pm25_per_tahun.values, marker='o', linestyle='-')
plt.title('Rata-rata PM2.5 berdasarkan Tahun')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata PM2.5')
plt.grid(True)

# Menambahkan label untuk setiap titik
for i, txt in enumerate(rata2_pm25_per_tahun.values):
    plt.text(rata2_pm25_per_tahun.index[i], txt, f'{txt:.2f}', ha='right', va='bottom')

plt.show()
```

##### Kesimpulan : 

Dari grafik terebut, didapatkan bahwa rata-rata PM2.5 pada Shunyi setiap tahunnya selalu berada diatas 40 mikrogram per meter kubik. 
Hal tersebut dapat diartikan bahwa kualitas udara di Shunyi buruk, karena menurut WHO batas ambang untuk PM2.5 adalah 15 mikrogram per meter kubik.

### Pertanyaan 2

#### Apakah terdapat pola pergerakan pada kualitas udara Shunyi setiap tahunnya berdasarkan PM10?

```python
rata2_pm10_per_tahun_bulan = df.groupby(['year', 'month'])['PM10'].mean()
rata2_pm10_per_tahun_bulan.index = rata2_pm10_per_tahun_bulan.index.map(lambda x: f'{x[0]}-{x[1]:02}')

plt.figure(figsize=(14, 6))
plt.plot(rata2_pm10_per_tahun_bulan.index, rata2_pm10_per_tahun_bulan.values, marker='o', linestyle='-')
plt.title('Rata-rata PM10 berdasarkan Bulan di Setiap Tahun')
plt.xlabel('Tahun dan Bulan')
plt.ylabel('Rata-rata PM10')
plt.xticks(rotation=45) 
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### Kesimpulan : 

Dari grafik tersebut, didapatkan bahwa terdapat pola pergerakan dimana pada bulan ke-12 rata-rata PM10 hampir selalu berada dibawah, kemudian rata-rata tersebut naik lagi hingga bulan ke-4

## Analisis Regresi

```python
import pandas as pd
import statsmodels.api as sm

# Membaca dataset dari file CSV
df = pd.read_csv("Shunyi_clean_data.csv")

# Memilih variabel independen (X) dan variabel dependen (y)
X = df[['TEMP']]  # Variabel independen: suhu udara
y = df['PM2.5']   # Variabel dependen: konsentrasi PM2.5

# Menambahkan konstanta ke variabel independen (X) untuk estimasi bias
X = sm.add_constant(X)

# Membuat model regresi
model = sm.OLS(y, X).fit()

# Mencetak ringkasan model
print(model.summary())
```

Dari summary tersebut didapatkan bahwa : 

- Model regresi yang dibangun menunjukkan bahwa terdapat hubungan yang signifikan antara variabel dependen (PM2.5) dan variabel independen (TEMP).

- Koefisien regresi untuk variabel TEMP adalah 0.3535, dengan nilai p yang sangat rendah (< 0.001), menunjukkan bahwa terdapat hubungan yang signifikan antara suhu udara dan konsentrasi PM2.5.

- Interval kepercayaan 95% untuk koefisien regresi TEMP tidak mencakup nol, menunjukkan bahwa kita cukup yakin bahwa efek suhu udara terhadap konsentrasi PM2.5 tidak nol.

#### Kesimpulan : 

Berdasarkan hasil ini, kita dapat menyimpulkan bahwa variabel PM2.5 bergantung pada variabel TEMP. Dengan kata lain, konsentrasi PM2.5 dipengaruhi oleh suhu udara, dan perubahan dalam suhu udara dapat memengaruhi perubahan dalam konsentrasi PM2.5. Meskipun hubungan ini signifikan secara statistik, nilai koefisien regresi yang rendah menunjukkan bahwa efek suhu udara mungkin tidak memiliki pengaruh yang besar dalam menjelaskan variasi dalam konsentrasi PM2.5.
